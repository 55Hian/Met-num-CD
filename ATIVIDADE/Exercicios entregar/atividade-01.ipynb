```python
import pandas as pd
import matplotlib.pyplot as plt



def atividade01():
    mortes_tendecia = []
    i = 0

    dados_csv = pd.read_csv("./cases-brazil-cities-time_2020.csv")

    inCityUsuario = input("digite a cidade desejado")

    dados_csv_por_cidade = dados_csv.loc[dados_csv["city"] == inCityUsuario]

    tempo_dados_csv_por_cidade = dados_csv_por_cidade["date"]

    novas_mortes_semana_cidade = dados_csv_por_cidade["newDeaths"]
    soma_mortes_semana_cidade = dados_csv_por_cidade["deaths"]
    novos_casos_semana_cidade = dados_csv_por_cidade["newCases"]
    soma_casos_semana_cidade = dados_csv_por_cidade["totalCases"]
    mortes_100k_habitantes_semana_cidade = dados_csv_por_cidade["deaths_per_100k_inhabitants"]
    casos_100k_habitantes_semana_cidade = dados_csv_por_cidade["deaths_per_100k_inhabitants"]
    mortes_relCasos_habitantes_semana_cidade = dados_csv_por_cidade["deaths_by_totalCases"]

    plt.plot(tempo_dados_csv_por_cidade, novas_mortes_semana_cidade, linestyle='-')
    plt.title("Novas mortes x Semana - Covid19 - " + inCityUsuario)
    plt.xlabel("Semanas")
    plt.ylabel("Novas mortes")
    plt.show()


    plt.plot(tempo_dados_csv_por_cidade, soma_mortes_semana_cidade, linestyle='-')
    plt.title("Mortes x Semana - Covid19 - " + inCityUsuario)
    plt.xlabel("Semanas")
    plt.ylabel("Mortes")
    plt.show()

    plt.plot(tempo_dados_csv_por_cidade, novos_casos_semana_cidade, linestyle='-')
    plt.title("Novos casos x Semana - Covid19 - " + inCityUsuario)
    plt.xlabel("Semanas")
    plt.ylabel("Novos Casos")
    plt.show()

    plt.plot(tempo_dados_csv_por_cidade, soma_casos_semana_cidade, linestyle='-')
    plt.title("Casos x Semana - Covid19 - " + inCityUsuario)
    plt.xlabel("Semanas")
    plt.ylabel("Casos")
    plt.show()

    plt.plot(tempo_dados_csv_por_cidade, mortes_100k_habitantes_semana_cidade, linestyle='-')
    plt.title("Mortes a cada 100 mil habitantes x Semana - Covid19 - " + inCityUsuario)
    plt.xlabel("Semanas")
    plt.ylabel("Mortes")
    plt.show()

    plt.plot(tempo_dados_csv_por_cidade, casos_100k_habitantes_semana_cidade, linestyle='-')
    plt.title("Casos a cada 100 mil habitantes x Semana - Covid19 - " + inCityUsuario)
    plt.xlabel("Semanas")
    plt.ylabel("Casos")
    plt.show()

    plt.plot(tempo_dados_csv_por_cidade, mortes_relCasos_habitantes_semana_cidade, linestyle='-')
    plt.title("Mortes com relação aos Casos x Semana - Covid19 - " + inCityUsuario)
    plt.xlabel("Semanas")
    plt.ylabel("Mortes")
    plt.show()

atividade01()



```
