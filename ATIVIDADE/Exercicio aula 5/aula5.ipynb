```python
import matplotlib.pyplot as plt

# Dados fornecidos
valores_x = [0, 1, 2, 3]
valores_fx = [9, 14, 12, 15]

# Número de pontos
n = len(valores_x)

# Calcular as somas necessárias
soma_x = sum(valores_x)
soma_fx = sum(valores_fx)
soma_x_quadrado = sum(x_i**2 for x_i in valores_x)
soma_x_fx = sum(x_i * fx_i for x_i, fx_i in zip(valores_x, valores_fx))

# Calcular os coeficientes da regressão linear
a = 3 #(n * soma_x_fx - soma_x * soma_fx) / (n * soma_x_quadrado - soma_x**2)
b = 6 # (soma_fx - a * soma_x) / n

# Criar a função g(x)
def g(x):
    return a * x + b

# Calcular os valores previstos
valores_fx_pred = [g(x_i) for x_i in valores_x]

# Calcular o erro médio quadrático
mse = sum((fx_i - g_x_i)**2 for fx_i, g_x_i in zip(valores_fx, valores_fx_pred)) / n

# Plotar os pontos e a função g(x)


plt.scatter(valores_x, valores_fx, label='Dados Originais')
plt.plot(valores_x, [g(x_i) for x_i in valores_x], color='red', label='Função Ajustada: g(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Ajuste de Curva Linear')
plt.legend()
plt.show()

print("Coeficientes da regressão linear (a, b):", (a, b))
print("Erro médio quadrático:", mse)
```
