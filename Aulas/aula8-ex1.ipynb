```python
import matplotlib.pyplot as plt

x = []
y = []
y.append(1)
x.append(0)
h = 0.1
i = 1
while i < 20:
    k1 = -x[i-1]*y[i-1]
    k2 = -(x[i-1]*y[i-1]+h*k1)
    x.append(x[i-1]+h)
    y.append(y[i-1]+h*(k1+k2)/2)
    print(f"(x(i):.3f)  (y[i]:.3f)")
    i = i + 1
    fig, ax = plt.subplots(figsize=(6, 6))
    plt.title("Exemplo: RK 2 ordem")
    plt.xlabel("x")
    plt.ylabel("y(x)")
    ax.plot(x,y, label = "curva")
    plt.show()    
```
